```python
import sys
import json

def _find(edge,node,n):
    list=edge[n]
    nl=[0,0,0]
    nl[2]=list[2]
    for i in range(0,31):
        if list[0]==node[i][0]:
            nl[0]=i
            break
    for i in range(0,31):
        if list[1]==node[i][0]:
            nl[1]=i
            break
    return nl

f = open("./hw4_data.json", 'r')
a=f.readline()
dic=json.loads(a)
nodes=dic["nodes"]
edges=dic["edges"]
lenn=len(nodes)
lene=len(edges)
arr=[]
for i in range(0,lenn):
    arr.append([])
    for j in range(0,lenn):
        arr[i].append(0)
for i in range(0,lene):
    nl=_find(edges,nodes,i)
    arr[nl[0]][nl[1]]=nl[2]
    arr[nl[1]][nl[0]]=nl[2]
for i in range(0,lenn):
    print(arr[i])
```

```python
for i in range(0,lenn):
    name=nodes[i][0]
    ll=[]
    for j in range(0,lenn):
        if arr[i][j]!=0:
            ll.append(nodes[j][0])
    print(name,':',ll)
```

```python
list1=[]
list2=[]
for i in range (0,lenn):
    list1.append(i)
    list2.append(nodes[i][1])
for i in range(0,5):
    for j in range(0,lenn-1-i):
        if list2[j]>list2[j+1]:
            list2[j],list2[j+1]=list2[j+1],list2[j]
            list1[j],list1[j+1]=list1[j+1],list1[j]
k=lenn-1
while k>=lenn-5:
    print(nodes[list1[k]][0])
    k-=1
```

```python
num=0
maxscore=0
for i in range(0,lenn):
    score=0
    for j in range(0,lenn):
        if arr[i][j]!=0:
            score+=1
    score+=nodes[i][1]
    if score>maxscore:
        maxscore=score
        num=i
print(nodes[num][0])
```

```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()
for i in range(0,lenn):
    G.add_node(nodes[i][0])
for i in range(0,lenn):
    for j in range(i+1,lenn):
        if arr[i][j]!=0:
            G.add_edge(nodes[i][0],nodes[j][0])

pos = nx.circular_layout(G)
nx.draw(G, pos, with_labels=True,font_size=8)
plt.show()
```
