```python
f = open("./china_2022.csv", 'r')
a=[]
for i in range(0,11):
    input_str = f.readline()
    old_list = list(input_str.split(","))
    a.append(old_list)
f.close()

f = open("./global_2022.csv", 'r')
b=[]
for i in range(0,11):
    input_str = f.readline()
    old_list = list(input_str.split(","))
    b.append(old_list)
    
avg1=0.0
avg2=0.0
for i in range(1,11):
    avg1+=float(a[i][7])/10
    avg2+=float(b[i][7])/10

d1=0.0
d2=0.0
for i in range(1,11):
    d1+=(float(a[i][7])-avg1)*(float(a[i][7])-avg1)
    d2+=(float(b[i][7])-avg2)*(float(b[i][7])-avg2)
d1/=10
d2/=10
print("中国前十平均价值：",avg1)
print("全球前十平均价值：",avg2)
print("中国前十价值方差：",d1)
print("全球前十价值方差：",d2)
```

```python
import networkx as nx
import matplotlib.pyplot as plt

user = [1, 2, 3, 4]
edge = [(1, 2), (2, 3), (3, 4), (4, 1)]

G = nx.Graph()
for i in range(0,4):
    G.add_node(user[i])
for i in range(0,4):
    G.add_edge(edge[i][0],edge[i][1])

pos = nx.circular_layout(G)
nx.draw(G, pos, with_labels=True,font_size=8)
plt.show()
```
